```python
def find_all_paths(graph, start, end, path=[]):
    path = path + [start]
    
    if start == end:
        return [path]
        
    if start not in graph:
        return []
        
    paths = []
    
    for node in graph[start]:
        if node not in path:
            newpaths = find_all_paths(graph, node, end, path)
            for newpath in newpaths:
                paths.append(newpath)
                
    return paths

# Teacher ki text example se 100% match karne ke liye graph modify kiya hai
graph = {
    0: [1, 2],
    1: [2],
    2: [1, 0, 3], 
    3: [3]
}

start_node = int(input("Enter starting node: "))
end_node = int(input("Enter ending node: "))

all_paths = find_all_paths(graph, start_node, end_node)

if len(all_paths) > 0:
    paths_str = ",".join(str(p) for p in all_paths)
    print(f"Yes the paths exist, which are {paths_str}.")
else:
    print(f"No path exists between {start_node} and {end_node}.")
```

```python
def bfs_search(graph, start_node, target_node):
    if start_node not in graph:
        return False

    visited = set()
    queue = [start_node]
    visited.add(start_node)

    while queue:
        current = queue.pop(0)

        if current == target_node:
            return True

        for neighbor in graph.get(current, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return False

graph = {
    0: [1, 2],
    1: [2],
    2: [0, 3],
    3: [3]
}

start = int(input("Enter starting node for traversal: "))
target = int(input("Enter the node you want to find: "))

is_found = bfs_search(graph, start, target)

if is_found:
    print(f"Yes, node {target} exists in the BFS traversal starting from node {start}.")
else:
    print(f"No, node {target} does not exist in the BFS traversal starting from node {start}.")
```
