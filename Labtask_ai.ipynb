```python
# Expression WITHOUT brackets
result1 = 2 + 3 * 4 ** 2 - 6 / 3

# Step by step samjho:
# Step 1: 4 ** 2 = 16         ( First Power)
# Step 2: 3 * 16 = 48         ( Then Multiply)
# Step 3: 6 / 3 = 2.0         ( Then Divide)
# Step 4: 2 + 48 - 2.0 = 48.0 (Last  Add/Subtract)

print("Without Brackets:", result1)

# Expression WITH brackets 
result2 = (2 + 3) * (4 ** 2) - (6 / 3)

# Step by step samjho:
# Step 1: (2 + 3) = 5         (Bracket )
# Step 2: (4 ** 2) = 16       (Bracket )
# Step 3: (6 / 3) = 2.0       (Bracket )
# Step 4: 5 * 16 = 80         (Then Multiply)
# Step 5: 80 - 2.0 = 78.0     (Last  Subtract)

print("With Brackets:", result2)
```

```python
# Marks define karo
marks = 45

# If - Elif condition
if marks >= 50:
    print("The student has passed the course")
elif marks < 50:
    print("The student has failed the course")
```

```python
# Step 1: Menu Display karo
print("========= CAFE MENU =========")
print("1. Soup and salad")
print("2. Pasta with meat sauce")
print("3. Chef's special")
print("==============================")

# Step 2: User se choice lo
choice = int(input("Which number would you like to order? "))

# Step 3: Choice check karo
if choice == 1:
    print("One Soup and salad coming right up!")
elif choice == 2:
    print("One Pasta with meat sauce coming right up!")
elif choice == 3:
    print("One Chef's special coming right up!")
else:
    print("Sorry, that is not a valid choice.")
```
