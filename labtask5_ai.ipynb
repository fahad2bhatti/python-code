```python
# ========== Vacuum Cleaner Agent ==========

print("=" * 50)
print("   VACUUM CLEANER AGENT SIMULATION")
print("=" * 50)

# Taking input from user for Room A
print("\nEnter condition for each room (Dirty / Clean):\n")
room_a = input("Enter condition of Room A: ")
room_b = input("Enter condition of Room B: ")

# Displaying entered values
print("\n" + "-" * 50)
print("You entered the following conditions:")
print(f"  Room A: {room_a}")
print(f"  Room B: {room_b}")
print("-" * 50)

# -------- Agent starts from Room A --------
print("\n>>> Agent starts from Room A <<<\n")

# Check Room A
print("Agent is in Room A...")
if room_a.lower() == "dirty":
    print("  Room A is Dirty → Action: CLEANING Room A ✅")
    room_a = "Clean"
    print(f"  Room A is now: {room_a}")
else:
    print("  Room A is Clean → Action: MOVE to Room B ➡️")

# Check Room B
print("\nAgent is in Room B...")
if room_b.lower() == "dirty":
    print("  Room B is Dirty → Action: CLEANING Room B ✅")
    room_b = "Clean"
    print(f"  Room B is now: {room_b}")
else:
    print("  Room B is Clean → Action: NO action needed ✅")

# Final Status
print("\n" + "=" * 50)
print("FINAL STATUS:")
print(f"  Room A: {room_a}")
print(f"  Room B: {room_b}")
print("  All rooms are CLEAN! Agent task completed! 🎉")
print("=" * 50)
```
