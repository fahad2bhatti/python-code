```python
graph = {
    0: [1, 2],
    1: [2],
    2: [0, 3],
    3: [3]
}

def dfs_paths(graph, start, goal, path=[]):
    path = path + [start]

    if start == goal:
        return [path]

    paths = []
    for neighbor in graph[start]:
        if neighbor not in path:
            new_paths = dfs_paths(graph, neighbor, goal, path)
            for p in new_paths:
                paths.append(p)

    return paths

# Example
start = 2
goal = 1

paths = dfs_paths(graph, start, goal)

if paths:
    print("Yes the paths exist, which are", paths)
else:
    print("No path exists")
```

```python
from collections import deque

graph = {
    0: [1, 2],
    1: [2],
    2: [0, 3],
    3: [3]
}

def bfs(graph, start, target):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()

        if node == target:
            return "Yes, node exists in graph traversal"

        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

    return "No, node does not exist in graph traversal"

# Example
start = 2
target = 1
print(bfs(graph, start, target))
```
