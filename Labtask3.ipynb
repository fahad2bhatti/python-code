```python
# Function to find the largest number from a list
def find_largest(numbers):
    largest = numbers[0]
    for num in numbers:
        if num > largest:
            largest = num
    return largest

# Function to get 10 numbers from user
def get_numbers():
    numbers = []
    print("Enter 10 numbers:")
    for i in range(1, 11):
        num = float(input(f"Enter number {i}: "))
        numbers.append(num)
    return numbers

# -------- Main Program --------
numbers = get_numbers()

# Display all entered numbers
print("\nYou entered the following numbers:")
for i in range(len(numbers)):
    print(f"Number {i+1}: {numbers[i]}")

result = find_largest(numbers)
print(f"\nThe largest number is: {result}")
```

```python
class basic_calc:
    # Constructor - values passed during instance creation
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    # Addition method
    def addition(self):
        result = self.num1 + self.num2
        print(f"{self.num1} + {self.num2} = {result}")
        return result

    # Subtraction method
    def subtraction(self):
        result = self.num1 - self.num2
        print(f"{self.num1} - {self.num2} = {result}")
        return result

    # Multiplication method
    def multiplication(self):
        result = self.num1 * self.num2
        print(f"{self.num1} x {self.num2} = {result}")
        return result

    # Division method
    def division(self):
        if self.num2 != 0:
            result = self.num1 / self.num2
            print(f"{self.num1} / {self.num2} = {result}")
            return result
        else:
            print("Error: Division by zero!")
            return None


# -------- Main Program --------
print("===== Basic Calculator =====")
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# Display entered numbers
print(f"\nYou entered the following numbers:")
print(f"First number: {a}")
print(f"Second number: {b}")

# Creating object (values passed with instance creation)
calc = basic_calc(a, b)

print("\nResults:")
calc.addition()
calc.subtraction()
calc.multiplication()
calc.division()
```
